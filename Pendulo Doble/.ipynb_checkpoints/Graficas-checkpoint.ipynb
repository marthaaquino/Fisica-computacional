```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib 
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.colors as mcolors
import matplotlib as mpl
```

```python
datos_fractal_irracional = pd.read_csv("diferencia_fractales.csv") #pd.read_csv("fractal_w(pi-8).csv")
datos_1 =  np.array(datos_fractal_irracional)
def rangos_func(data):
    #lista de donde buscará la diferenicia de tiempo
    rangos_lista =  [(0,50),(50,100),(100,500),(500,1000)]
    #número de conjuntos en los que se dividirán los datos
    N = len(rangos_lista)
    delta = 0
    #la matriz donde se anexarán los valores
    datos_reescalado = np.zeros([len(data[:,1]), len(data[1,:]) ])
    #para analizar todos los datos
    for i in range(0,len(data[:,1])):
        for j in range(0,len(data[:,1])):
            for k in range(N):
                #aquí comparamos
                if rangos_lista[k][0]< data[i,j] <rangos_lista[k][1]:
                    datos_reescalado[i,j] = k 
                #elif rangos_lista[k][0]< data[i,j] <rangos_lista[k][1]:
                #    datos_reescalado[i,j] = delta*k 
    
                
    datos_reescalado_2 = np.zeros([len(data[:,1]), len(data[1,:]) ])
    #por alguna extraña razón el fractal aparece como reflejado, entonces aquí cambiamos eso
    for j in range(0,len(datos_reescalado[1,:])):
        for i in range(0,len(datos_reescalado[:,1])):
            datos_reescalado_2[i,j] = datos_reescalado[i,-j-1]
            
    return datos_reescalado_2


puntos = rangos_func(datos_1)
datos = puntos

rangos_strings =  [(0,50),(50,100),(100,500),(500,1000)]
rangos = ["(0,50)","(50,100)","(100,500)","(500,1000)"]
# Numero de conjuntos de puntos que se convertiran en graficas sobrepuestas.
num_graficas = len(puntos)
# Color de degradado de colores. Se toma la paleta y se pide la cantidad de colores.
cmap_colors = sns.color_palette("RdBu_r",  n_colors=len(rangos_strings))
# cmap_colors = sns.color_palette("light:#AE0000_r", n_colors=len(aaa))  # Otro ejemplo

# - - - - Tamaño de la imagen - - - -


figsize = 8
espacio_barra_color = 2.8
fig, ax = plt.subplots(dpi=100,  figsize=(figsize+espacio_barra_color, figsize))
marker_size = 2 # Con el tamaño se debe ajustar este parametro tambien. Solo acepta enteros.

title_size = 30
x_ticks_size = y_ticks_size = 25
delta = 1.9e-2 # Espacio entre borde e imagen.

csfont = {'fontname':'Times New Roman'}

###plt.title("$\mathregular{\omega ≈}$ 990 [Hz]", size=title_size, **csfont)
ax.set_xlabel("$\mathregular{θ_1 [rad]}$", size=20, **csfont)
ax.set_ylabel("$\mathregular{θ_2 [rad]}$", size=20, **csfont)

plt.xticks(fontsize=x_ticks_size, **csfont)
plt.yticks(fontsize=y_ticks_size, **csfont)

#etiquetas de los ejes
labels = ['-3',"-2", '-1','0', '1','2','3']
#Posiciones de esas etiquetas, esto está muy rustico, podría modificarse
posicion_menos3 = -627*3/np.pi+627
posicion_menos2 = -627*2/np.pi+627
posicion_menos1 = -627*1/np.pi+627
posicion_0 = 627
posicion_1 = 1254-(-627*1/np.pi+627)
posicion_2 = 1254-(-627*2/np.pi+627)
posicion_3 = 1254-(-627*3/np.pi+627)
#las anexamos a un arreglo
ax.set_xticks([posicion_menos3,posicion_menos2,posicion_menos1,posicion_0, posicion_1, posicion_2, posicion_3])
#que las ponga en la gráfica
ax.set_xticklabels(labels)

#Lo mismo para las etiquetas en y
posicion_menos3_y = 1257*3/np.pi
posicion_menos2_y =1257*2/np.pi+628*1/np.pi
posicion_menos1_y =1257*1/np.pi+628*2/np.pi
posicion_0_y = 629
posicion_1_y =  629 + -1257/(2*np.pi)
posicion_2_y =  629 + -1257/np.pi
posicion_3_y = -1257*3/np.pi+1257
ax.set_yticks([posicion_menos3_y,posicion_menos2_y,posicion_menos1_y,posicion_0_y, posicion_1_y, posicion_2_y, posicion_3_y])
ax.set_yticklabels(labels)


sns.set(context = 'paper')
# - - - - Sin grid y el color del grid - - - -
sns.set_style("ticks", {'axes.grid' : False})


# Description of each category : 
labels = np.array(rangos_strings)
len_lab = len(labels)
# prepare normalizer
# El más 1 que se pone, es porque si no, no toma toda la lista de labels que le damos
# Entonces eso lo corrige.
norm = matplotlib.colors.BoundaryNorm( [i for i in range(0, len_lab+1)] , len_lab, clip=True)
# lambda x es una funcion que no es externa, es una asigacion local para fmt.
fmt = matplotlib.ticker.FuncFormatter(lambda x, pos: labels[norm(x)])

# Prepare bins for the normalizer
norm_bins = np.array([i for i in range(len(rangos_strings))])
# Esto es para que se centren a la mitad, cada rango.
diff = np.ones(len_lab )
tickz = norm_bins[:] + diff / 2


# - - - - Mostrar barra de color - - - - -
barra_color = plt.cm.ScalarMappable(cmap=ListedColormap(cmap_colors),
                           norm=plt.Normalize(vmin=0, vmax=len(labels) ))
barra_color._A = []

cmap = plt.get_cmap('RdBu_r', 4)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

cbar = plt.colorbar(sm, format=fmt, ticks=tickz) # Posicion de cada label
                    

cbar.set_label('Diferencias en los tiempos de giro. ',  size=20,  **csfont)
cbar.ax.tick_params(labelsize=15) 

plt.imshow(datos, cmap = "RdBu_r", aspect = "auto")#, Interpolation  = "lanczos")


#norm = mpl.colors.Normalize(vmin=0, vmax=1)
  
plt.savefig("diferencia.png")
plt.show() #cubehelix #Spectral




```
