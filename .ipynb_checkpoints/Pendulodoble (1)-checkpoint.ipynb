```python
import numpy as np
import matplotlib.pyplot as plt

w = 1.0
```


Ecuaciones adimensionalizadas:

\begin{align*}
\dot \theta_1= \frac{p_1-p_2\cos(\theta_1-\theta_2)}{(1+\sin^2(\theta_1-\theta_2))}
\end{align*}


\begin{align*}
\dot \theta_2= \frac{2p_2-p_1\cos(\theta_1-\theta_2)}{(1+\sin^2(\theta_1-\theta_2))}
\end{align*}

\begin{align*}
\dot p_1=-2\omega^2\sin \theta_1-A+B
\end{align*}

\begin{align*}
\dot p_2=-2\omega^2\sin \theta_2+A-B
\end{align*}
donde:
\begin{align*}
A=\frac{p_1p_2\sin (\theta_1-\theta_2)}{(1+\sin^2(\theta_1-\theta_2))}
\end{align*}

\begin{align*}
B=\frac{p_1^2+2p_2^2-p_1p_2\cos (\theta_1-\theta_2)}{2(1+\sin^2(\theta_1-\theta_2))^2}
\end{align*}


```python
# Ecuaciones

def pendulo(t1, t2, p1, p2):
    dcos = np.cos(t1 - t2)
    dsin = np.sin(t1 - t2)
    D = 1 + dsin**2
    A = (p1 * p2 * dsin) / D    
    B = (p1**2 + 2 * p2**2 - p1 * p2 * dcos) / (2 * D**2)
    return (p1 - p2 * dcos) / D, (2 * p2 -  p1 * dcos) / D, -2 * w * np.sin(t1) - A + B, -2 * w * np.sin(t2) + A - B

# Runge Kutta 4to Orden
def rk4(t1, t2, p1, p2, h):
    h=0.1
    # k1
    k1_t1, k1_t2, k1_p1, k1_p2 = pendulo(t1, t2, p1, p2)
    
    # k2
    k2_t1, k2_t2, k2_p1, k2_p2 = pendulo(t1 + h * k1_t1 / 2, t2 + h * k1_t2 / 2, p1 + h * k1_p1 / 2, p2 + h * k1_p2 / 2)
    
    # k3
    k3_t1, k3_t2, k3_p1, k3_p2 = pendulo(t1 + h * k2_t1 / 2, t2 + h * k2_t2 / 2, p1 + h * k2_p1 / 2, p2 + h * k2_p2 / 2)
    
    # k4
    k4_t1, k4_t2, k4_p1, k4_p2 = pendulo(t1 + h * k3_t1, t2 + h * k3_t2, p1 + h * k3_p1, p2 + h * k3_p2)
    
    # Nuevos valores
    t1_sol = t1 + h * (k1_t1 + 2 * k2_t1 + 2 * k3_t1 + k4_t1) / 6
    t2_sol = t2 + h * (k1_t2 + 2 * k2_t2 + 2 * k3_t2 + k4_t2) / 6
    p1_sol = p1 + h * (k1_p1 + 2 * k2_p1 + 2 * k3_p1 + k4_p1) / 6
    p2_sol = p2 + h * (k1_p2 + 2 * k2_p2 + 2 * k3_p2 + k4_p2) / 6
    
    return t1_sol, t2_sol, p1_sol, p2_sol

#comprobando rk4, esta parte es importante no sólo para ver que todo funciones bien,
#sino que también te permite entender como funciona.
#A continuación te dejo una prueba de como implementarlo para 
#que lo pienses, de una forma similar se resuelve para el tiempo de giro.
#Solo es pensar como hacer que se detenga cuando detecte que dio un giro
puntos = 1000
#Tiempo para el cual se obtendrán las soluciones
tiempo = np.linspace(0,15,puntos)
h = (tiempo[-1]-tiempo[0])/puntos
solt1 = np.zeros(puntos)
solt1[0] = 0.01
solt2 = np.zeros(puntos)
solt2[0] = 0.01
solp1 = np.zeros(puntos)
solp1[0] = 0
solp2 = np.zeros(puntos)
solp2[0] = 0
for i in range(1, puntos):
    soluciones = rk4(solt1[i-1],solt2[i-1],solp1[i-1],solp2[i-1],h)
    solt1[i] = soluciones[0]
    solt2[i] = soluciones[1]
    solp1[i] = soluciones[2]
    solp2[i] = soluciones[3]
plt.plot(tiempo,solt1)
```

```python

#Resolvemos RK4, si observas, esta f es lo que implementé arriba
#sin embargo inicie definiendo el tamaño de t, t1A,t2A, p1A y p2A
#Esta implementación puede hacer que tu código corra más rápido ya que
#Es más rápido para la computadora llena un espacio de un arreglo con cierta
#información que "pegarsela" al fina

def solu(t1_0, t2_0):
    
    # Parámetros
    t_0 = 0
    t_f = 15
    h = 0.1
    
    #Condiciones iniciales
    t = [t_0]
    t1A = [t1_0]
    t2A = [t2_0]
    p1A = [0]
    p2A = [0]
    
    
    c=0 #Contador
    for i in np.arange(t_0 + h, t_f + h, h):
    
        sol =  rk4(t1A[c], t2A[c], p1A[c], p2A[c], h)
        t.append(i)
        t1A.append(sol[0])
        t2A.append(sol[1])  
        p1A.append(sol[2])  
        p2A.append(sol[3])

        c=c+1

    return t, t1A, t2A

#Después de haber definido tu función es prudente probarla, así si tiene
#errores puede detectarse antes

soluciones_assad = solu(0.01, 0.01)
plt.plot(soluciones_assad[0], soluciones_assad[1])
```

```python
N = 100 #puntos para malla ¡cuidado! estamos pensando en una malla cuadrada, dado el proceso de abajo
#esto me dice que NxN son los números en la malla
d_pi = 1 / N  # Delta pi



info = [] #Creamos arreglo para hacer después el mapa de calor
#¿es necesario ese delta pi? en ambos for. Que posición tendría tu pendulo en eso

for t1_0 in np.arange(-np.pi, np.pi + d_pi, d_pi):
 
    for t2_0 in np.arange(0, np.pi + d_pi, d_pi): # 0 a pi porque sólo queremos dos cuadrantes

        if 2 * np.cos(t1_0) - np.cos(t2_0) <= 1:  # Mínima energía
            continue
        soluc = solu(t1_0, t2_0)
        tiempo = soluc[0]
        t1_tiempo = soluc[1]
        t2_tiempo = soluc[2]
        
    
                
#pequeño detalle, que pasa cuando no cumple la condición? inmediatamente no hace nada, porque no guarda información de que para
#las condiciones con las que no gira 
```

```python
#propuesta
N = 25
d_pi = 1 / (N**2) 

t_final = 10  #tiempo final 
info = np.zeros([N,N]) 
#Con esto haremos un arreglo matricial, para el cual cada entrada tiene su correspondientes condiciones iniciales, que serán
#guardadas en este arreglo
malla_condiciones_iniciales = np.zeros([N,N])
#En vez de contar a partir de los angulos, contaremos cada elemento en tu malla, para ello nos auxileremos de este arreglo
Angulos = np.linspace(-np.pi,np.pi,N)
for i in range(0,N):
    t1_0 = Angulos[i]
    for j in range(0,N):
        t2_0 = Angulos[j]
        
        #Vemos si cumplen la condición la que no giran
        if 2*np.cos(t1_0)+np.cos(t2_0)>1:
            #guardamos en la matriz de info que para esta configuración de ángulos no gira
            info[j,i] = t_final + 10
        #ahora si para el tiempo final se obtiene las soluciones, para ello voy a copiar tu codigo
        else: 
            soluc = solu(t1_0, t2_0)
            tiempo = soluc[0]
            t1_tiempo = soluc[1]
            t2_tiempo = soluc[2]
        #aqui no habría que inicia desde 0, sino de un poco más adelante, para descartas los elementos de frontera donde
        #alguno de los dos pendulos se encuentra en pi al tiempo 0
        for k in np.arange(10, len(tiempo)):
            epsilon = 0.05
            if abs(t1_tiempo[k]-np.pi) <= epsilon or abs(t2_tiempo[k]-np.pi) <= epsilon:
                t_g = tiempo[k]  #tiempo de giro
                info[j,i] = (t_g)
                break
                
#pequeño detalle, que pasa cuando no cumple la condición? inmediatamente no hace nada, porque no guarda información de que para
#las condiciones con las que no gira 
```

```python
plt.imshow(info)
plt.colorbar()
```

```python
# Mapa de calor

grid_x, grid_y = np.linspace(-np.pi, np.pi, N), np.linspace(-np.pi, np.pi, N)
grid_x, grid_y = np.meshgrid(grid_x, grid_y)
grid_z = np.zeros((N, N))

for t_g, t1_0, t2_0 in info:
    x_idx = int((t1_0 + np.pi) / (2 * np.pi / (N - 1)))
    y_idx = int((t2_0 + np.pi) / (2 * np.pi / (N - 1)))
    grid_z[x_idx, y_idx] = t_g


plt.imshow(grid_z, extent=[-np.pi, np.pi, -np.pi, np.pi], origin='lower', aspect='auto', cmap='viridis')
plt.colorbar(label='$t_g$')
plt.xlabel('$t1_0$')
plt.ylabel('$t2_0$')
plt.title('Heatmap')
plt.show()

```
